```python
def get_box_region(mesh, padding_added):
    """
    Return the region for applying refinement on the mesh
    """
    # Get true boundaries of the mesh
    west, south = mesh.origin[:2]
    east = west + mesh.hx[0] * mesh.hx.size
    north = south + mesh.hy[0] * mesh.hy.size
    # Remove the padding added to the region
    return (west + padding, east - padding, south + padding, north - padding)
```

## Build synthetic body

### Define active cells and density array

```python
active_cells = np.ones(mesh.n_cells, dtype=bool)
```

```python
density = np.zeros(mesh.n_cells)
```

Get center of the mesh

```python
easting_c = (region[0] + region[1]) / 2
northing_c = (region[2] + region[3]) / 2
easting_c, northing_c
```

Add single rectangle as synthetic source

```python
half_width = 20
west = easting_c - half_width
east = easting_c + half_width
south = northing_c - half_width
north = northing_c + half_width
bottom = -40
top = -10


centers = mesh.gridCC
easting, northing, upward = centers[:, 0], centers[:, 1], centers[:, 2]
inside = (
    (easting > west) & (easting < east)
    & (northing > south) & (northing < north)
    & (upward > bottom) & (upward < top)
)
density[inside] = 0.5
```

Plot the mesh

```python
# Define plotting map to fill with nans the inactive cells
plotting_map = SimPEG.maps.InjectActiveCells(mesh, active_cells, np.nan)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

ax1, ax2 = axes

maxabs = vd.maxabs(density)

for ax, dim in zip(axes, "x y".split()):
    tmp, = mesh.plot_slice(
        plotting_map * density,
        ax=ax,
        normal=dim,
        clim=(-maxabs, maxabs),
        pcolor_opts=dict(cmap="RdBu_r"),
    )
    ax.set_aspect("equal")

cb = plt.colorbar(tmp, ax=axes, orientation="horizontal", label="g/cc", pad=0.08)
cb.ax.set_xscale("linear")
plt.show()
```

```python
SLICE_LOC = -20

tmp, = mesh.plot_slice(
    plotting_map * density,
    slice_loc=SLICE_LOC,
    normal="z",
    clim=(-maxabs, maxabs),
    pcolor_opts=dict(cmap="RdBu_r"),
)
plt.grid()
plt.gca().set_aspect("equal")
plt.colorbar(tmp)
plt.show()
```

## Compute forward model of synthetic 

```python
grid_coords = vd.grid_coordinates(region, spacing=10, extra_coords=10)
```

```python
print(f"Number of data points: {grid_coords[0].size}")
```

Define gradiometry survey

```python
coordinates = tuple(c.ravel() for c in grid_coords)
survey = create_gravity_survey(coordinates)
```

Define wires mappings

```python
inject_active_cells_map = SimPEG.maps.InjectActiveCells(
    mesh, active_cells, np.nan
)
wires = SimPEG.maps.Wires(
    ("density", inject_active_cells_map.nP),
)
```

```python
gravity_simulation = potential_fields.gravity.simulation.Simulation3DIntegral(
    survey=survey,
    mesh=mesh,
    rhoMap=wires.density,
    ind_active=active_cells,
)
```

```python
g_z = gravity_simulation.dpred(density)
```

Add gaussian noise to the synthetic data

```python
percentage = 1/100
rng = np.random.default_rng(42)

std = percentage * vd.maxabs(g_z)
g_z += rng.normal(loc=0, scale=std, size=g_z.size)
```

```python
shape = grid_coords[0].shape
g_z = g_z.reshape(shape)
```

```python
plt.pcolormesh(*grid_coords[:2], g_z)
plt.title("g_z")
plt.gca().set_aspect("equal")
plt.gca().grid()
plt.colorbar()
plt.show()
```

## Run sparse inversion

```python
# Define uncertainties for gravity data
percentage = 1 / 100
g_z_std = percentage * vd.maxabs(g_z)
```

```python
# Define gravity data object
gravity_data = create_gravity_data(
    coordinates, g_z.ravel(), g_z_std
)
```

Build mapping for the mesh

```python
n_active_cells = active_cells.sum()
model_map = SimPEG.maps.IdentityMap(nP=n_active_cells)
```

Create reference model

```python
reference_model = np.zeros(n_active_cells)
```

Define the data misfit and the regularizations

```python
data_misfit = SimPEG.data_misfit.L2DataMisfit(
    data=gravity_data, simulation=gravity_simulation
)
data_misfit
```

Build the regularization

```python
regularization = SimPEG.regularization.Sparse(
    mesh,
    active_cells=active_cells,
    mapping=model_map,
)
regularization.norms = [0, 1, 1, 1]
```

Define the optimization algorithm

```python
optimization = SimPEG.optimization.ProjectedGNCG(
    maxIter=50,    # max number of iterations
    lower=-1.0,    # bounds for the model parameters (in g/cc)
    upper=1.0,     # 
    maxIterLS=20,  # maximum number of iterations for line search
    maxIterCG=10,  # maximum number of iterations for the Conjugate Gradient
    tolCG=1e-3,    # tolerance for conj grad
)
optimization
```

Define inverse problem

```python
inverse_problem = SimPEG.inverse_problem.BaseInvProblem(
    data_misfit, regularization, optimization
)
```

Define directives

```python
# Define a starting value for the trade-off parameter (beta)
starting_beta = SimPEG.directives.BetaEstimate_ByEig(beta0_ratio=1)

# IRLS
update_IRLS = SimPEG.directives.Update_IRLS(
    f_min_change=1e-4,
    max_irls_iterations=30,
    coolEpsFact=1.5,
    beta_tol=1e-2,
)

# Define the fractional decrease in beta and the number of
# Gauss-Newton solves for each beta value.
beta_schedule = SimPEG.directives.BetaSchedule(coolingFactor=3, coolingRate=1)

# Update the preconditionner if it is model dependent
update_jacobi = SimPEG.directives.UpdatePreconditioner()

# Set a stopping criteria for the inversion
# target_misfit = SimPEG.directives.TargetMisfit(chifact=0.5)

# Add sensitivity weights
sensitivity_weights = SimPEG.directives.UpdateSensitivityWeights(everyIter=False)

# Save results after every iteration
save_output = SimPEG.directives.SaveOutputEveryIteration(save_txt=False)

# Save model after every iteration
save_model = SimPEG.directives.SaveModelEveryIteration()
save_model.directory = str(intermediate_models_dir)
```

```python
# The directives are defined as a list.
directives = [
    update_IRLS,
    sensitivity_weights,
    starting_beta,
    beta_schedule,
    update_jacobi,
    # target_misfit,
    save_output,
    save_model,
]
```

Define inversion

```python
inversion = SimPEG.inversion.BaseInversion(inverse_problem, directives)
```

### Run inversion

```python
start = time.time()
inverted_model = inversion.run(reference_model)
end = time.time()
print(f"Elapsed time: {(end - start) / 60:.2f} min")
```

## Plot the model

```python
def plot_model(mesh, model, vmin=None, vmax=None, cmap="viridis", label=None):
    from matplotlib.gridspec import GridSpec

    if vmin is None:
        vmin = np.min(model)
    if vmax is None:
        vmax = np.max(model)

    kwargs = dict(clim=(vmin, vmax), pcolor_opts=dict(cmap=cmap))

    # Create fig and axes
    fig = plt.figure(figsize=(16, 6))
    gs = GridSpec(nrows=2, ncols=2, width_ratios=[1.4, 1])
    axes = [
        fig.add_subplot(gs[0, 0]),
        fig.add_subplot(gs[1, 0]),
        fig.add_subplot(gs[:, 1]),
    ]
    # Vertical slices
    for ax, dim in zip(axes[:2], ["x", "y"]):
        slice_loc = None
        (tmp,) = mesh.plot_slice(
            model,
            ax=ax,
            normal=dim,
            slice_loc=slice_loc,
            **kwargs,
        )
    # Horizontal slice
    (tmp,) = mesh.plot_slice(
        model,
        ax=axes[-1],
        slice_loc=SLICE_LOC,
        normal="z",
        **kwargs,
    )
    # Extra stuff
    if label is None:
        label = ""
    plt.colorbar(tmp, ax=axes, label=label)
    for ax in axes:
        ax.set_aspect("equal")
        ax.grid()
```

```python
maxabs = vd.maxabs(density)
plot_model(mesh, density, vmin=-maxabs, vmax=maxabs, cmap="RdBu_r", label="g/cc")
plt.show()

maxabs = vd.maxabs(inverted_model)
plot_model(mesh, inverted_model, vmin=-maxabs, vmax=maxabs, cmap="RdBu_r", label="g/cc")
plt.show()
```
